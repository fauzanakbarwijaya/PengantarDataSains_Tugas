```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
file_path = 'titanic.csv'

print(f"--- 2. Load & EDA: {file_path} ---")
df = pd.read_csv(file_path)

print("Data 5 baris pertama (head):")
print(df.head())
print("\n")

print("Info Tipe Data & Cek Awal Missing Value (info):")
df.info()
print("\n")

print("Statistik Deskriptif (describe):")
print(df.describe())
print("\n")


print("--- 3. Deteksi Missing Values ---")
missing_values = df.isnull().sum()

print("Jumlah missing values per kolom (sebelum imputasi):")
print(missing_values[missing_values > 0])
print("\n")


print("--- 4. Imputasi Missing Values ---")

median_age = df['age'].median()
print(f"Nilai median 'age' yang akan digunakan: {median_age}")

df['age'] = df['age'].fillna(median_age)

print("Imputasi kolom 'age' selesai.\n")


print("--- Verifikasi Hasil Imputasi ---")
print("Cek ulang missing values setelah 'age' diisi:")
final_missing = df.isnull().sum()

print(final_missing[final_missing > 0])
print("\n")

print(f"Jumlah missing value di 'age' sekarang: {df['age'].isnull().sum()}")


output_filename = 'titanic_cleaned.csv'
df.to_csv(output_filename, index=False)
print(f"Data bersih telah disimpan ke: {output_filename}")
```
